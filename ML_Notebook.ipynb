```python
x = np.linspace(0,2*np.pi, 1000)
```

```python
plt.plot(x, y)
plt.grid()
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine Function")
plt.show()
```

```python
plt.plot(x, np.cos(x))
plt.grid()
plt.xlabel("X")
plt.ylabel("cos(x)")
plt.title("Cosine Function")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x-values avoiding division by zero (or extremely small values)
x = np.linspace(-3 * np.pi, 3 * np.pi, 1000)

# Calculate y = csc(x) as 1 / sin(x)
y = 1 / np.sin(x)

# Replace values near asymptotes with NaN to prevent Matplotlib from drawing connecting lines
y[np.abs(y) > 10] = np.nan

# Plotting the cosecant graph
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', linewidth=2, label='y = csc(x)')

# Add vertical asymptotes
for k in range(-3, 4):
    plt.axvline(x=k * np.pi, color='red', linestyle='--', alpha=0.6)

# Formatting the plot
plt.xlabel("x (radians)", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.ylim(-5, 5)
plt.axhline(0, color='black', linewidth=1)

plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.show()
```
