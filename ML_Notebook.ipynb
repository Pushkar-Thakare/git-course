```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
x = np.linspace(0,2*np.pi, 1000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.grid()
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine Function")
plt.show()
```

```python
plt.plot(x, np.cos(x))
plt.grid()
plt.xlabel("X")
plt.ylabel("cos(x)")
plt.title("Cosine Function")
plt.show()
```

```python
plt.plot(x, np.tan(x))
plt.grid()
plt.xlabel("x")
plt.ylabel("tan(x)")
plt.title("Tangent Function")
plt.show()
```

```python
k = np.linspace(-5, 5, 10000)
plt.plot(k, np.tanh(k))
plt.grid()
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Hyperbolic Tangent Function")
plt.show()
```

```python
k = np.linspace(-5, 5, 10000)
plt.plot(k, np.sinh(k))
plt.grid()
plt.xlabel("x")
plt.ylabel("sinh(x)")
plt.title("Hyperbolic Sine Function")
plt.show()
```

```python
k = np.linspace(-5, 5, 10000)
plt.plot(k, np.cosh(k))
plt.grid()
plt.xlabel("x")
plt.ylabel("cosh(x)")
plt.title("Hyperbolic Cosine Function")
plt.show()
```
